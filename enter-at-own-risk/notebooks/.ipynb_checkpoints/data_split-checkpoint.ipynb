```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
```

```python
# loading the dataset with labels
data = pd.read_pickle('../data/final_set.pkl')
# also loading the Bag of Words dataset
data_bow = pd.read_pickle('../data/data_set_bow.pkl')
```

```python
data.tail()
```

```python
data.drop(data.tail(1).index, inplace=True)
```

```python
Y = data['Labels']
```

```python
Y
```

```python
X = data_bow
```

```python
X.shape
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
len(X_test)
```

```python
len(X_train)
```

```python
# Using Gradient Boosting Classifier to train the model
from sklearn.ensemble import GradientBoostingClassifier as GBC
```

```python
classifier = GBC(n_estimators=100, learning_rate=0.1, max_depth=3)
```

```python
classifier.fit(X_train, Y_train)
```

```python
predictions = classifier.predict(X_test)
```

```python
predictions[:10]
```

```python
# Let's check out the accuracy 
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy = accuracy_score(Y_test, predictions)
```

```python
accuracy
```

```python
report = classification_report(Y_test, predictions)
```
